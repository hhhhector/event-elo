```python
import pandas as pd
```

```python
event_history = pd.read_json('../data/event_history.json')
to_add = pd.read_excel('../data/to_add.ods').T
```

```python
for index, row in to_add.iterrows():
    if index != 'Event':
        event = index
        date = row.iloc[0]
        player_count = row.iloc[1]
        players = list(row.iloc[2:].dropna())
        event_history.loc[len(event_history)] = (pd.Series({'event' : event, 'date' : date, 'player_count' : player_count, 'players' : players}))
```

```python
event_history = event_history.sort_values(by = 'date')
event_history['date'] = event_history['date'].astype(str)
```

```python
event_history.to_json('../data/event_history.json', orient='records', indent=4)
```
